Задача - необходимо решить диффур:

y" + pxcos(y) = 0

0 <= x <= 1

y'(0) = 0

y(1) = 0

p = {1, 4, 7, 25, 50, 100}

Для решения задачи используется метод стрельбы, точку подбирал руками

```python
import numpy as np              
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
h = 0.0001
N = 10000

ylast = 0.0
dy0 = 0.0
```

```python
def y_next(x_cur, y_prev, y_cur, p):
    return 2 * y_cur - y_prev - p * x_cur * np.cos(y_cur) * (h ** 2)
```

```python
def solve(y0, p):
    y = np.zeros(N, dtype = float)
    y[0] = y0
    y[1] = y[0] + dy0 * h

    for i in range(1, N - 1):
        y[i + 1] = y_next(h * i, y[i - 1], y[i], p)
        
    return y
```

```python
p = 1
y0 = 0.165

y = solve(y0, p)
print(y[N - 1], ylast)
```

```python
x = np.linspace(0, (N - 1) * h, N)

plt.figure(figsize = [8, 5])
plt.plot(x, y)
plt.title(label = 'p = 1')
plt.xlabel("x")
plt.ylabel("y")
plt.tight_layout() 
plt.grid()
plt.show()
```
